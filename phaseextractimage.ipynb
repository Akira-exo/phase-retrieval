```python
#define MODULE to convert an RGB image to its grayscale version
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
# convert read image to grayscale
i0=rgb2gray(img); 
plt.imshow(i0, cmap='gray')
plt.show()
```

```python
#convert nd array to 2d list for manipulation
i1= i0.tolist()
```

```python
f1=np.fft.fftn(i1); #take fourier transform of the object ( outputs a complex 2d list)

mag1= abs(f1); # take the magnitude of the image

s=np.log(1+np.fft.fftshift(f1)); # shifts the frequency of the obtained FTT symmetrically
      
phase1=np.angle(f1); # take the phase profile of the image

r1=np.fft.ifftshift(np.fft.ifftn(mag1)); # recreate object using magnitude information only

r2=np.fft.ifftn(np.exp(1j*phase1)); #recreate object using phase information only
```

```python
#module to convert complex number 2d list to an rgb image

from matplotlib.colors import hsv_to_rgb
def Complex2HSV(z, rmin, rmax, hue_start):
    # get amplitude of z and limit to [rmin, rmax]
    amp = np.abs(z)
    amp = np.where(amp < rmin, rmin, amp)
    amp = np.where(amp > rmax, rmax, amp)
    ph = np.angle(z, deg=1) + hue_start
    # HSV are values in range [0,1]
    h = (amp -rmin) / (rmax - rmin)
    v = (ph % 360) / 360
    s = 0.85 * np.ones_like(h)
    
    return hsv_to_rgb(np.dstack((h,s,v)))
```

```python
s_RGB= Complex2HSV(s,0.0,0.5,90) #convert the complex valued array into a RGB image
#(https://stackoverflow.com/questions/17044052/mathplotlib-imshow-complex-2d-array) 

plt.imshow(rgb2gray(s_RGB), cmap='gray') #RGB to grayscale convert and display
```
